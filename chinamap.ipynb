```python
import pandas as pd
import csv
import os
from pyecharts.charts import Map
from pyecharts import options as opts 
path = os.path.dirname(os.getcwd())+'/Pyecharts_plot/selleradrs_china.csv'
path
```

```python
proList = []
with open(path) as f:
    reader = csv.reader(f)
    for row in reader:
        pro_dict = dict()
        pro_dict['province'] = row[0]
        pro_dict['seller_num'] = row[1]
        proList.append(pro_dict)
proList
```

```python
province_df=pd.DataFrame(proList)
province_df
```

```python
c=Map(opts.InitOpts(width='1000px',height='800px'))
province_name=[str(x) for x in province_df.province]
province_seller_num=[int(x) for x in province_df.seller_num]

gen=[list(z) for z in zip(province_name,province_seller_num)]

#颜色配置
pieces=[
    {'max':1,'color':'#FFFFF0'},
    {'min':1,'max':9,'color':'#FFE0E0'},
    {'min':10,'max':49,'color':'#FEC0C0'},
    {'min':50,'max':99,'color':'#FD9090'},
    {'min':100,'max':149,'color':'#FC6060'},
    {'min':150,'max':199,'color':'#FB3030'},
    {'min':200,'color':'#DD0000'},
    
]
c.add('全国卖家分布',gen,'china')
c.set_global_opts(visualmap_opts=opts.VisualMapOpts(is_piecewise=True,pieces=pieces))
c.render_notebook()
c.render()
```
